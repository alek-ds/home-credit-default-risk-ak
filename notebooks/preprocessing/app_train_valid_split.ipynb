```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from pathlib import Path
import sys
```

```python
app_df = pd.read_csv(r"..\..\data\raw\application_train.csv")
print(f"Shape of app_df: {app_df.shape}")
app_df.head(10)
```

```python
X, y = app_df.drop(columns=['TARGET']), app_df['TARGET']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

print(f"X_train: {X_train.shape}")
print(f"X_valid: {X_valid.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_valid: {y_valid.shape}")
```

```python
X_train.head()
```

Since `sklearn.model_selection.train_test_split()` perserves `Pandas` indicies we can safely join them.

```python
assert X_train.index.equals(y_train.index)
assert X_valid.index.equals(y_valid.index)

app_train_df = X_train.join(y_train)
app_valid_df = X_valid.join(y_valid)
```

```python
PROJECT_PATH = Path.cwd().parent.parent
DATA_SAVE_PATH = PROJECT_PATH / 'data' / 'interim'

if str(DATA_SAVE_PATH) not in sys.path:
    sys.path.insert(0, str(DATA_SAVE_PATH))

DATA_SAVE_PATH.mkdir(parents=True, exist_ok=True)
```

```python
app_train_df.to_csv(DATA_SAVE_PATH / 'app_train.csv', index=False)
print(f"app_train_df saved to: DATA_SAVE_PATH\\app_train.csv")

app_valid_df.to_csv(DATA_SAVE_PATH / 'app_valid.csv', index=False)
print(f"app_valid_df saved to: DATA_SAVE_PATH\\app_vaild.csv")
```
